```python
import sys
#sys.path.append("..")
import jax
import jax.random as jr
import numpy as np
import equinox as eqx # pytree utilities
from rnn_jax.cells import ElmanRNNCell, ClockWorkRNNCell
from rnn_jax.layers import RNNEncoder
import matplotlib.pyplot as plt
import timeit
```

```python
rnn = RNNEncoder(ElmanRNNCell(10, 64, key=jr.key(0)))
cwrnn = RNNEncoder(ClockWorkRNNCell(10, [16, 16, 16, 16], [2, 4, 8, 16], nonlinearity=jax.nn.relu, key=jr.key(0)))
```

```python
x = jr.uniform(jr.key(1), (1000, 10))
jit_rnn = eqx.filter_jit(rnn)
jit_rnn(x)
jit_cwrnn = eqx.filter_jit(cwrnn)
jit_cwrnn(x)
%timeit jit_rnn(x)
%timeit jit_cwrnn(x)
```

```python
x = jr.uniform(jr.key(1), (256, 5000, 10))
rnn = RNNEncoder(ElmanRNNCell(10, 1024, key=jr.key(0)))
cwrnn =RNNEncoder(ClockWorkRNNCell(10, [1024//4]*4, [4, 16, 64, 256], nonlinearity=jax.nn.relu, key=jr.key(0)))
vmap_rnn = eqx.filter_vmap(rnn)
vmap_rnn(x)
vmap_cwrnn = eqx.filter_vmap(cwrnn)
vmap_cwrnn(x)
%timeit vmap_cwrnn(x)
%timeit vmap_cwrnn(x)
```
